```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats 
import pandas as pd
import geopandas as gpd
import cartopy.crs as ccrs
from matplotlib.lines import Line2D
from scipy.stats import linregress
```

```python

# Define the base folder and path to the eBI results CSV
base_folder = r"C:\Users\Feifei\Box\BR_remote_sensing\hma_rivers\AmuDarya_Kerki\rivgraph"
eBI_path = os.path.join(base_folder, 'eBI_results.csv')


# Read the eBI results data
eBI_data = pd.read_csv(eBI_path)

# Group the eBI data by year and calculate the annual average
eBI_annual_avg = eBI_data.groupby('Year')['eBI'].mean()

# Generate a list of all years (assuming you're analyzing from 1985 to 2023)
all_years = np.arange(1985, 2023 + 1)

# Reindex the DataFrame to include all years, filling missing years with NaN
eBI_annual_avg = eBI_annual_avg.reindex(all_years)

# Drop NaN values for regression and plotting
eBI_annual_avg_no_nan = eBI_annual_avg.dropna()
years = eBI_annual_avg_no_nan.index
eBI_values = eBI_annual_avg_no_nan.values

# Perform linear regression using scipy's linregress
slope, intercept, r_value, p_value, std_err = linregress(years, eBI_values)

# Generate the regression line
regression_line = slope * years + intercept

# Plotting
fig, ax = plt.subplots(figsize=(9, 4))

# Plot the annual averaged eBI values
ax.plot(years, eBI_values, label='Average eBI (Annual)', marker='o', linestyle='-', color='tab:blue', markersize=10)

# Plot the regression line
ax.plot(years, regression_line, label=f'Regression Line (slope={slope:.2f}, $R^2$={r_value**2:.2f})', color='tab:red', linestyle='--')

# Set x-ticks for every 4th year (adjust this if you want different intervals)
year_ticks = all_years[::4]  # Select every 4th year for x-ticks
ax.set_xticks(year_ticks)  # Set these years as the x-ticks
ax.set_xticklabels(year_ticks)  # Set the labels
ax.tick_params(axis='x', labelsize=15)  # x-tick labels
ax.tick_params(axis='y', labelsize=15)  # y-tick labels

# Set labels and title
ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Reach-averaged eBI', fontsize=15)
plt.title('Amu Darya', fontsize=15)

# Add legend and show plot
plt.legend()
plt.tight_layout()  # Adjust layout to fit
plt.show()
```
